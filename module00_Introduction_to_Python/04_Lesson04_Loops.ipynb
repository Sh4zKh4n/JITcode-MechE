Alright, so three steps, and now we've hit a wall.  We *could* continue through and plan out all of Rover's actions through the maze, but that is not very efficient, and it wouldn't work if our mazed changed somehow.  So let's figure out what's in the maze in front of us.  To do this, we can bring up the value of the location *infront* of us in the maze.  Let's do that now.

```python
maze1=np.loadtxt(fname='maze1.csv', delimiter=',')
Rover = np.array([7,7])
pointing = np.array([0,-1])
maze1[infront[0],infront[1]]
```

The way that we've made the maze is this: If we've never been there, and it's safe terrain, the location in the maze is valued zero.  If we have been there, and it's safe terrain, the location of the maze is valued one.  If it isn't safe, we consider the value of the location to be 3 or 5, depending on if it is a wall or an edge.  For our purposes, a three and five can be considered the same thing.  

We know that our function takes a string and returns a string.  We also already know that it works, from the last lesson.  Let's make a function called 'danger' that will determine if the place in front of us is safe, using the maze and the index of place we're considering moving towards.  Using the maze makes sense; think about it like your sensor looking at the environment.

```python
def danger(maze,index):
    if maze[index[0],index[1]] == 3 or maze[index[0],index[1]] == 5:
        return 'hazard'
    else:
        return 'safe'
```

Let's make sure it works.  If we give it the value of infront that we know, we should get the word 'safe' back.  If we give it the index [0,0](the top-left corner of the maze), we should get the word 'hazard' back.  Let's try that.

```python
print(danger(maze1,infront))
print(danger(maze1,[0,0]))
```

Ok. so we've got our function that tells us what to do, but we don't have anything that actually turns us right.  It's actually really easy.  We can use a *Rotation Matrix*.  Essentially, this is a nifty little tool that just rotates directions.  To turn right 90 degrees, the matrix looks like this:

```python
rturn = np.array([[0,1],[-1,0]])
print(rturn)
```

To turn using this matrix is even easier.  we just use the dot() function, which multiplies the way we're facing by the matrix, *rotating* it in the process. Check this out.

```python
pointing = np.array([0,-1])
print('before rotation we face:')
print(pointing)
pointing = rturn.dot(pointing)
print('after rotation we face:')
print(pointing)
```

Thinking about this in terms of indexing makes sense.  If we were facing left [0,-1], then turn right, we would be facing up [-1,0].  Neat, huh?  We can use this to make short, efficient code.

Let's bring it together to take 10 steps without crashing through a wall.  To do this we'll need to first find the location in front of us, see if it's dangerous, decide whether or not to turn right, and then, finally, either turn right or step forward.  Let's try it.

```python
#this is just so we can create an animation of the path. you can ignore it for now.
maze1=np.loadtxt(fname='maze1.csv', delimiter=',')
mazehist = maze1.copy()

Rover = np.array([7,7])
pointing = np.array([0,-1])
for i in range(0,10):
    infront = Rover + pointing
    is_safe = danger(maze1,infront)
    command = student_func(is_safe)
    if command == 'go forward':
        Rover = infront
        maze1[Rover[0],Rover[1]] = 1
    if command == 'turn right':
        pointing = rturn.dot(pointing)
    
    #this line build up frames for our animation
    mazehist = np.dstack((mazehist,maze1))     
```

And, once we've run that, let's animate it so we can watch!

```python
fig = plt.figure(figsize=(8,4)) #defining the size of our figure

#this section builds our animation
ims = []
for i in range(mazehist.shape[2]):
    im = plt.imshow(mazehist[:,:,i], cmap=plt.cm.Reds, interpolation='nearest')
    ims.append([im])
anim = animation.ArtistAnimation(fig, ims, interval=100)

#this one displays it
display_animation(anim, default_mode='once')
```

How about that!  Now he's moving.  There is the problem, however, of not knowing (without counting) how many steps it will take rover to get to the goal.  Let's use a *While loop* to remedy the situation.  A while loop will continue to do something until a condition is no longer true.  We use it all the time in real life; we do homework until it is done, we cook food until it is ready. Think of a while loop as considering the state of something instead of how many times we'd like to do something.  This makes it perfect for our maze; we want to navigate *while* we aren't at our destination([1,4] on the maze].  

To test whether or not we're at our destination, we can use the *np.array_equal()* function, which just returns a true value if an array is equal to another and a false value if they are not.

There is a certain danger of a while loop, though.  If you aren't careful, you can create an *infinite* loop.  If there is no way that the computer can ever make it out of the loop, it will keep doing it until you force the program to stop!  We'd like to avoid that, so we've made a dummy 'count' variable that will *break* us out of the loop if we get stuck.  This is generally pretty good practice.  

```python
#this is just so we can create an animation of the path. you can ignore it for now.
maze1=np.loadtxt(fname='maze1.csv', delimiter=',')
mazehist = maze1.copy()

destination = np.array([1,4])
Rover = np.array([7,7])
pointing = np.array([0,-1])

count = 0

while not (np.array_equal(Rover,destination)):
    infront = Rover + pointing
    is_safe = danger(maze1,infront)
    command = student_func(is_safe)
    if command == 'go forward':
        Rover = infront
        maze1[Rover[0],Rover[1]] = 1
    if command == 'turn right':
        pointing = rturn.dot(pointing)
    if count == 100:
        print('stuck in the loop')
        break
    count = count + 1
    
    #this line build up frames for our animation
    mazehist = np.dstack((mazehist,maze1))     
```

Here's the animation:

```python
fig = plt.figure(figsize=(8,4)) #defining the size of our figure

#this section builds our animation
ims = []
for i in range(mazehist.shape[2]):
    im = plt.imshow(mazehist[:,:,i], cmap=plt.cm.Reds, interpolation='nearest')
    ims.append([im])
anim = animation.ArtistAnimation(fig, ims, interval=100)

#this one displays it
display_animation(anim, default_mode='once')
```
