```python
import Image
import math
import numpy as np

from IPython.display import display
from IPython.display import Image as iPyImage

import scipy
from scipy.ndimage import filters as filters

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Load_Image(image_name):
    img = Image.open('resources/'+image_name)
    return np.array(img)

def Draw_Image(data):
    plt.imshow(data, cmap=plt.cm.binary, interpolation='none')
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
tmp_img = Load_Image("lena.png");
Draw_Image(tmp_img)
```

```python
def Remove_Color(img):
    output = np.zeros(img.shape[:2])
    for y in range(len(img)):
        for x in range(len(img[y])):
            output[y][x] = int(img[y][x][0]*0.2+img[y][x][1]*0.7+ img[y][x][2]*0.1)
    return output
```

```python
black_white = Remove_Color(tmp_img)
Draw_Image(black_white)
```

```python
def Blur(img, power):
    return filters.gaussian_filter(img, power)
            
```

```python
blur_img = Blur(black_white,2)
Draw_Image(blur_img)
```

```python
def Sobel_Transform(img):
    horizontal = filters.sobel(img,1)
    vertical   = filters.sobel(img,0)
    
    magnitude = np.zeros(horizontal.shape)
    angle     = np.zeros(horizontal.shape)
    rangle    = np.zeros(horizontal.shape)
    
    for y in range(len(horizontal)):
        for x in range(len(horizontal[y])):
            Gx = horizontal[y][x]
            Gy = vertical[y][x]
            
            mag  = math.sqrt(Gy**2+Gx**2)
            ang  = math.atan2(Gx,Gy)
            rang = math.atan2(Gx,Gy)
            while ang < 0:
                ang += math.pi*2
            ang = round(ang/math.pi*4)
            while ang >= 4:
                ang -= 4
            magnitude[y][x] = mag
            angle[y][x]     = ang
            rangle[y][x]    = rang
    
    return (magnitude, angle, rangle)
    
```

```python
mag, ang, rang = Sobel_Transform(blur_img)
Draw_Image(mag)
Draw_Image(ang)

```

```python
def Canny_Transform(mag, ang, thresh):
    output = np.zeros(mag.shape)
    for y in range(1,len(mag)-1):
        for x in range(1,len(mag[y])-1):
            if ang[y][x] == 3 and mag[y][x-1]<mag[y][x]>mag[y][x+1]:
                output[y][x] = 1
            if ang[y][x] == 0 and mag[y-1][x+1]<mag[y][x]>mag[y+1][x-1]:
                output[y][x] = 1
            if ang[y][x] == 1 and mag[y-1][x]<mag[y][x]>mag[y+1][x]:
                output[y][x] = 1
            if ang[y][x] == 2 and mag[y+1][x+1]<mag[y][x]>mag[y-1][x-1]:
                output[y][x] = 1
            if mag[y][x] < thresh:
                output[y][x] = 0
    return output
```

```python
canny_img = Canny_Transform(mag,ang,50)
Draw_Image(canny_img)
```

```python
def Detect_Lines(img, showImages = False):
    if (showImages):
        Draw_Image(img)
    bnw            = Remove_Color(img)
    if (showImages):
        Draw_Image(bnw)
    blr            = Blur(bnw, 4)
    if (showImages):
        Draw_Image(blr)
    mag, ang, rang = Sobel_Transform(blr)
    if (showImages):
        Draw_Image(mag)
        Draw_Image(ang)
        Draw_Image(rang)
    can            = Canny_Transform(mag,ang, 40)
    if (showImages):
        Draw_Image(can)
    return mag, ang, rang, can
```

```python
ball = Load_Image("ball.jpg")
mag, ang, rang, can = Detect_Lines(ball, True)
```

```python
def Draw_MB_Line(m, b, img):
    for y in range(len(img[0])):
        x = m*y+b
        if 0<x<len(img):
            img[x][y] += 1
    return img
    
```

```python
def Hough_Circle(can, rang):
    output = np.zeros(can.shape)
    for y in range(2,len(can)-2):
        for x in range(2,len(can[y])-2):
            if (can[y][x] == 1):
                m = 1./(math.tan(rang[y][x]) or .01)
                b = y-m*x
                Draw_MB_Line(m, b, output)
    return output
            
```

```python
lines = Hough_Circle(can, rang)
Draw_Image(lines)
```

```python
def Find_Center(img):
    max_x, max_y, max_val = 0,0,0
    for y in range(len(img)):
        for x in range(len(img[y])):
            if img[y][x] > max_val:
                max_x   = x
                max_y   = y
                max_val = img[y][x]
    return max_x, max_y
```

```python
max_x, max_y = Find_Center(lines)
print max_x, max_y
```

```python
def Find_Radius(img, c_x, c_y):
    size = math.sqrt(img.shape[0]**2 + img.shape[1]**2)
    accumulator = np.zeros(size)
    for y in range(len(img)):
        for x in range(len(img[y])):
            if img[y][x] == 1:
                dist = math.sqrt( (c_x-x)**2 + (c_y-y)**2 )
                if (dist < 10): continue
                accumulator[int(dist)] += 100./dist
    
    max_val = 0
    max_rad = 0
    for i in range(len(accumulator)):
        if accumulator[i] > max_val:
            max_val = accumulator[i]
            max_rad = i
    return max_rad
```

```python
rad = Find_Radius(can, max_x, max_y)
print rad
```

```python
def Draw_Circle(img, c_x, c_y, rad, color):
    output = img.copy()
    for theta in np.linspace(0, math.ceil(2*math.pi), math.ceil(rad*math.pi*2)):
        x = c_x + rad*math.cos(theta)
        y = c_y + rad*math.sin(theta)
        try:
            output[y][x] = color
        except IndexError:
            pass
    
    output[c_y+1][c_x+1] = color
    output[c_y+1][c_x+0] = color
    output[c_y+1][c_x-1] = color
    output[c_y+0][c_x+1] = color
    output[c_y+0][c_x+0] = color
    output[c_y+0][c_x-1] = color
    output[c_y-1][c_x+1] = color
    output[c_y-1][c_x+0] = color
    output[c_y-1][c_x-1] = color
    
    return output
```

```python
circle_ball = Draw_Circle(ball, max_x, max_y, rad, (255, 0, 255))
Draw_Image(circle_ball)
```

```python
def Find_Circle_and_Center(img, showImg = False):
    mag, ang, rang, can = Detect_Lines(img, showImg)
    lines = Hough_Circle(can, rang)
    max_x, max_y = Find_Center(lines)
    
    if showImg:
        Draw_Image(lines)
        rad = Find_Radius(can, max_x, max_y)
        circle_ball = Draw_Circle(img, max_x, max_y, rad, ((255, 0, 255, 255) if img.shape[2] == 4 else (255, 0, 255)) )
        Draw_Image(circle_ball)
    
    return max_x, max_y
```
