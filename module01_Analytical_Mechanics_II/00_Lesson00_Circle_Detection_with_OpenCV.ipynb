```python
import Image
import math
import numpy as np
import cv2
import cv2.cv as cv

from IPython.display import display
from IPython.display import Image as iPyImage

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Load_Image(image_name):
    img  = cv2.imread('resources/'+image_name, 0)
    img  = cv2.medianBlur(img,3)
    return img

def Draw_Image(data):
    plt.imshow(data, cmap=plt.cm.binary, interpolation='none')
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
tmp_img = Load_Image("lena.png");
Draw_Image(tmp_img)
edges = cv2.Canny(tmp_img, 150, 200)
Draw_Image(edges)
```

```python
img = Load_Image('ball00/frame7.png')
Draw_Image(img)
```

```python
def Find_Center(img):
    circles = cv2.HoughCircles(img,cv.CV_HOUGH_GRADIENT,1,20,param2=10)
    if circles is not None and len(circles) >= 1:
        return circles[0,:][0][:2]
    return None
```

```python
output = np.ones((256,256,3))
points = []

for i in range(19):
    tmp_img = Load_Image('ball00/frame'+str(i)+'.png')
    center = Find_Center(tmp_img)
    if center is not None:
        points.append(center)
        cv2.circle(output,tuple(center),2,(0,0,255),3)
    
Draw_Image(output)
    
```

```python
dx  = 0   # dx/dy
dy  = []  # dy/dt
d2y = 0   # d^2y/dt^2

for p in range(len(points)-1):
    dx += points[p+1][0]-points[p][0]
    dy.append(points[p+1][1]-points[p][1])
dx = float(dx) / (len(points)-1)

for p in range(len(dy)-1):
    d2y += dy[p+1]-dy[p]
d2y = float(d2y) / (len(dy)-1)

dy0 = -(points[1][1] - points[0][1])

v0 = math.sqrt(dx**2+dy0**2)
t  = math.atan2(dy0,dx)

print "Initial Velocity: %f pixels/frame"%v0
print "    Horizontal  : %f pixels/frame"%dx
print "    Vertical    : %f pixels/frame"%dy0
print "Initial Angle   : %f degrees"%((t*180)/math.pi)
print "Acceleration    : %f pixels/frame^2"%-d2y
```

####For reference, the image file was generated with a horizontal velocity of 10 (vs. 11 calculated), a vertical velocity of 45 (vs. 38 calculated [although we did ignore the first and last frames so this is expected]), and an acceleration of 5 (vs. 5.1 calculated)
