```python
import Image
import math
import numpy as np

import cv2
from cv2 import cv as cv

from IPython.display import display
from IPython.display import Image as iPyImage

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Load_Image(image_name):
    tmp = Image.open("resources/"+image_name)
    return np.array(tmp)

def Draw_Image(data):
    plt.imshow(data, cmap=plt.cm.binary, interpolation='none')
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
tmp1 = Load_Image("blur/frame7.png")
Draw_Image(tmp1)
tmp2 = Load_Image("blur/frame6.png")
Draw_Image(tmp2)
```

```python
def Image_Diff(img1, img2):
    
    out = np.ones(img1.shape[:2], dtype="uint8")
    for y in range(len(img1)):
        for x in range(len(img1[y])):
            a1 = img1[y][x]
            a2 = img2[y][x]
            r = a1[0] - a2[0] if a1[0] > a2[0] else a2[0] - a1[0]
            g = a1[1] - a2[1] if a1[1] > a2[1] else a2[1] - a1[1]
            b = a1[2] - a2[2] if a1[2] > a2[2] else a2[2] - a1[2]
            
            mag = .7*r+.2*g+.1*b
            
            if mag < 10:
                mag = 0
                
            out[y][x] = mag

    return out#cv2.medianBlur(cv2.fromArray(out),3)
```

```python
boo = Image_Diff(tmp1,tmp2)
print boo
Draw_Image(boo)
```

```python
def Find_Line(img, show = False):
    blur  = cv2.medianBlur(img,3)
    lines = cv2.HoughLines(blur,1,np.pi/180,10)
    
    if lines == None:
        return
    
    r,t = lines[0][0]
    m = -1./math.tan(t)
    b = r/math.sin(t)

    if show:
        copy = np.copy(blur)
        for x in range(len(blur[0])):
            y = m*x+b
            if 0<y<len(blur):
                copy[y][x] = 20
        Draw_Image(copy)
        
    return m,b
```

```python
images = range(18)
for i in range(1,17):
    images[i] = Load_Image("blur/frame"+str(i)+".png")
    
for i in range(1,16):
    diff = Image_Diff(images[i],images[i+1])
    Find_Line(diff, True)
```
