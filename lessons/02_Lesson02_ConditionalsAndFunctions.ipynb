How should we navigate? Hmm... We'll start with something simple. Assume we have a sensor on the front of our rover which detects if there is a hazard in front of it.  In this case, let's turn right.  Here's a *function* which will do this.

```python
def student_func(front_sensor):
    if front_sensor == 'hazard':
        return 'turn right'
    else:
        return 'go forward'
```

Cool! we've made our first function. Take note of the syntax describing how we make a function.  We start with the word "def", which means "define", and end the line with a colon.  After that, every indented line is included in the function.  Once you stop indenting, you've stopped writing the function.  Easy as that!  

`student_func` is the *name* of the function.  Function names follow the same rules as variables.  They can't start with numbers and they can't contain any special characters besides `_` characters, which are commonly used instead of spaces.  In the parenthesis are the parameters of the function.  If you want to think of a function like a recipe, the parameters are the function's raw ingredients.  In this case, the function accepts the parameter `front_sensor`.  Don't worry too much about the contents of the indented lines, also called the *body* of the function.  We'll describe what they're doing in a moment.

Functions are nice because they encapsulate something we want to do more than once; it saves us typing and makes our code more readable.  Watch this. 

```python
student_func('hazard')
```

```python
student_func('clear')
```

Looks like our function works!  Let's talk a bit more about how, and why.  Inside our function, you see our very first use of a *conditional* statement.  Conditional statements are really useful; they let our programs make decisions based on some data they receive.  to use a conditional, we use a double equals sign, `==`, to tell the program that it should find out if two things are equal.  There are other things the computer can check, but let's focus on this for right now. 

*if* the comparison proves true, then everything indented underneath the if statement will be performed.  *else* we do what is under the else statement.  Let's see how we navigate the grid.

Imagine that we've attached a sensor to Rover's left side.  We can turn left if there isn't anything there when we hit rocks, right?  Let's write that function!

```python
def student_func(front_sensor, left_sensor):
    if (front_sensor == 'hazard') and (left_sensor == 'hazard'):
        return 'turn right'
    if (front_sensor == 'hazard') and (left_sensor != 'hazard'): 
        return 'turn left'
    else:
        return 'go forward'
```

Cool! Now we'll turn left whenever we can, and right when we can't.  Let's look at how that works.

See the `and` keyword?  This means that we can have our control logic check two things at the same time. The first case is when the rover gets itself next to two walls, one in front, and one on his left.  The second means there is only a wall in front of it.  

See the `!=` test?  Weird, huh?  the `!=` inverts the test to mean *is not equal to*. That's right, we can check negatives, too.  Decision structures can be pretty powerful, as I'm sure you're starting to see.
